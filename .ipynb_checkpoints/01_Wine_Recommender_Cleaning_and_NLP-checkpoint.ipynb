# Data cleaning

## Import libraries and load data

```python
import pandas as pd
import numpy as np

from bs4 import BeautifulSoup

import matplotlib.pyplot as plt
import seaborn as sns

import re

import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords

%matplotlib inline

wine = pd.read_csv('data/winemag-data-130k-v2.csv')
```

## Functions

```python
def headers_to_lowercase (df):
    df.columns = df.columns.str.lower()
    return df

def text_to_lowercase (df):
    df = df.applymap(lambda s: s.lower() if type(s) == str else s)
    return df   

def drop_columns (df, list_of_columns_to_drop):    
    df.drop(columns = list_of_columns_to_drop, inplace=True)
    return df

def change_col_names (df, column_name_dict):
    df.rename(columns=column_name_dict, inplace= True)
    return df

def replace_nans_in_price_with_winery_median (df):
    df['price_in_usd'] = df['price_in_usd'].fillna(df.groupby('winery')['price_in_usd'].transform('median'))
    return df

def drop_rows_with_nans (df, col_list_for_row_drop):
    for i in col_list_for_row_drop:
        df.drop(df[df[i].isna()].index, inplace=True)
    return df

def change_float_to_int(df):
    df['price_in_usd']=df['price_in_usd'].astype(int)
    return df

def drop_duplicates(df):
    df.drop_duplicates(inplace=True,ignore_index=True)
    return(df)
```

## Variables for data cleaning

```python
col_list_for_row_drop = ['designation','region','price_in_usd']

list_of_columns_to_drop = ['unnamed: 0','region_2','taster_name','taster_twitter_handle']

column_name_dict = {'price':'price_in_usd','region_1':'region','title':'wine_name'}
```

## Cleaning pipeline

```python
wine = (wine
.pipe(headers_to_lowercase)
.pipe(text_to_lowercase)
.pipe(drop_columns, list_of_columns_to_drop)
.pipe(change_col_names, column_name_dict)
.pipe(replace_nans_in_price_with_winery_median)
.pipe(drop_rows_with_nans, col_list_for_row_drop)
.pipe(change_float_to_int)
.pipe(drop_duplicates)
)
```

# Using NLTK functions on the wine dataframe

## Functions to iterate over the description column

```python
def clean_text(string):
    
    '''this function cleans the text from all non-alphanumeric symbols, double whitespaces and digits'''
    
    string = re.sub(r'\W+',' ',string) # Replace everything non-alpahnumeric by ' '
    string = re.sub(r'\s+',' ',string) # Replace one or more whitespaces by  ' '
    string = re.sub(r'\d+',' ',string) # Replace one or more digits by  ' '
    
    return string


def tokenize_and_lemmatize_string(string):
    
    '''this function uses the NLTK functions TOKENIZER and LEMMATIZER.
    tokenizing separates the text into separate strings where each string is one word, 
    lemmatizer changes the words to a more basic grammatical form.'''

    lem = WordNetLemmatizer()

    tokenized_string = nltk.word_tokenize(string)
    
    lemmatized_tokens = []
    
    for word in tokenized_string:
        lem.lemmatize(word)
        lemmatized_tokens.append(word) 
        
    return lemmatized_tokens


def pos_tagger(lemmatized_tokens):
    
    '''POS-tagger is a function from the NLTK library, which tags each word with the part of speach tag,
    so recognizes if the word is a verb, noun, adjective and gives it the corresponding tag.
    the outcome is a list of tuples, where index 0 is the word, and index 1 is the tag.'''
    
    tagged_words = nltk.pos_tag(lemmatized_tokens)
    
    return tagged_words
    

def remove_unnecessary_pos_tags(tagged_words):
    
    '''in this project we decided that the most relevant words that describe wines the best would be 
    nouns and adjectives, so we removed all other words from the descriptions using this function.'''
    
    selected_tags = ['JJ','NN','NNP']
    
    selected_words = []

    for tup in tagged_words:
        if tup[1] in selected_tags:
            selected_words.append(tup[0])
        else:
            continue
    
    return selected_words


    
def remove_stopwords(selected_words):
    
    '''this function removes stopwords, which very often get tagged with a wrong tag, 
    the list of stopwords is included in the NLTK library.'''
    
    stop_words = list(stopwords.words('english')) 
    
    selected_words_1 = []

    for word in selected_words:
        if word not in stop_words:
            selected_words_1.append(word)
        else:
            continue
        
    return selected_words_1  
    
    

def remove_short_words(selected_words_1):
    
    '''this function removes all words that are shorter than 3 letters.'''
    
    selected_words_2 = []

    for word in selected_words_1:
        if len(word) > 2:
            selected_words_2.append(word)
        else:
            continue

    return selected_words_2
```

## Apply functions on dataframe

```python
# clean text with regex
wine['description'] = list(map(clean_text, wine['description']))

# tokenize and lemmatize text
wine['description'] = list(map(tokenize_and_lemmatize_string, wine['description']))

# pos-tagging
wine['description'] = list(map(pos_tagger, wine['description']))

# remove all tags except nouns and adjectives 
wine['description'] = list(map(remove_unnecessary_pos_tags, wine['description']))

# removing stopwords
wine['description'] = list(map(remove_stopwords, wine['description']))

# removing short words
wine['description'] = list(map(remove_short_words, wine['description']))
```

## Creating new columns with scent profiles

```python
for i in list_of_column_names:
    
    wine[i] = "0" 
    
wine
```

## Looking for keywords from dictionary in wine descriptions

We look for keywords from the dictionary in each wine description and add a 1 in the corresponding column if keyword from category is in the description. 

```python
for i in range(len(wine['description'])):

    for key, value in scent_dict.items():
        for word in value:

            if word in wine.loc[i,'description']:
                wine.loc[i,key] = '1'
                
            else:
                continue
```
